```python
import pandas as pd

dados_pib = pd.read_csv('Dados/pib-regioes.csv')

dados_pib.head()
```

```python
# Calculando a quantidade de valores nulos
nan_count = dados_pib.isna().sum().sum()
nan_count
```

```python
dados_indice_aneel = pd.read_csv('Dados/indice-satisfacao-aneel.csv')
dados_indice_aneel.head()
```

```python
nan_count = dados_indice_aneel.isna().sum().sum()
proporcao_nulos = nan_count / dados_indice_aneel.size
proporcao_nulos
```

```python
dados_pib.rename(columns={"Grande Região": "Regiao", "Produto Interno Bruto a preços correntes (Mil Reais)": "PIB"}, inplace=True)
dados_pib = dados_pib[["Regiao", "Ano", "PIB"]]
regioes_pib = dados_pib.groupby('Regiao')['PIB'].mean().reset_index()
regioes_pib
```

```python
dados_indice_aneel.rename(columns={"DescricaoCategoria": "Regiao"}, inplace=True)

dados_indice_aneel = dados_indice_aneel[["NumAno", "DscClassificacao", "Regiao", "MdaIndicadorQualidade"]]

join = dados_indice_aneel.merge(regioes_pib, how='inner', on='Regiao')
join.query('NumAno >= 2006 & NumAno <=2020')
```
